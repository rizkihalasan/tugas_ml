```python
import random

class KMeans:
    
    def __init__(self, **kwargs):
        if 'n_clusters' in kwargs:
            self.n_clusters = int(kwargs['n_clusters'])
        if 'max_iter' in kwargs:
            self.max_iter = int(kwargs['max_iter'])
        else :
            self.max_iter = 300
        self.cluster_centers_ = []
        self.labels_ = []
        self.n_iter_ = 0
        
    
    def euclidian_distance(self, A, B):
        res = 0
        for i in range(len(A)):
            res += (A[i]-B[i]) ** 2
        return res ** 0.5
    
    
    def fit(self, data):
        # ambil n buah random
        self.cluster_centers_ = random.sample(data, self.n_clusters)
        dist = [[] for x in range(len(data))]
        print(self.cluster_centers_)
        stop = False
        self.n_iter_ = 0
        
        while (not stop) :
            self.n_iter_ += 1
            print("\niterasi ", self.n_iter_)
            print("centroid ", self.cluster_centers_)
            #hitung jarak ke seluruh point dari centroid tsb
            for i in range(len(data)):
                dist[i] = [];
                for j in range(len(self.cluster_centers_)):
                    dist[i].append(self.euclidian_distance(data[i],self.cluster_centers_[j]))
            print("jarak",dist)
            self.labels_ = []
            for i in range(len(dist)):
                self.labels_.append(dist[i].index(min(dist[i])))
            print("cluster",self.labels_)

            total_dist = [[0 for y in range(len(data[0]))] for x in range(self.n_clusters)]
            cluster_data_count = [0 for x in range(self.n_clusters)]
            for i  in range(len(self.labels_)) :
                for j in range(len(data[i])) :
                    total_dist[self.labels_[i]][j] += data[i][j]
                cluster_data_count[self.labels_[i]] += 1
            print("total dist", total_dist)
            print("cluster total", cluster_data_count)
            for i in range(len(total_dist)):
                for j in range(len(total_dist[i])) :
                    total_dist[i][j] = total_dist[i][j]/cluster_data_count[i]
                    
            # hitung error
            error = 0
            for i  in range(len(total_dist)):
                error += self.euclidian_distance(total_dist[i], self.cluster_centers_[i])
            print("error",error)
            if (error < 0.01) or (self.max_iter and (self.n_iter_ > self.max_iter)) :
                stop = True
            else :
                self.cluster_centers_ = total_dist
    
    def predict(self, Y) :
        result = []
        dist = [[] for x in range(len(Y))]
        #hitung jarak ke seluruh point dari centroid tsb
        for i in range(len(Y)):
            dist[i] = [];
            for j in range(len(self.cluster_centers_)):
                dist[i].append(self.euclidian_distance(Y[i],self.cluster_centers_[j]))
        print("jarak",dist)
        self.labels_ = []
        for i in range(len(dist)):
            result.append(dist[i].index(min(dist[i])))
        print("cluster",result)
        return result
            
        
        
hehe = KMeans(n_clusters=3)
A = [[2, 10],[2, 5],[8, 4],[5, 8],[7, 5],[6,4],[1,2],[4,9]]
hehe.fit(A)
hehe.predict([[1.5,3.5],[4,9]])
    
    
```
